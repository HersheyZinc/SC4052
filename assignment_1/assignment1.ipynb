```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from tcp_algorithms import plot, aimd, dctcp, cubic

ITERATESMAX = 100  # You can adjust this value as needed 
C = 12  # Adjust C value as needed 
```

## AIMD

```python
x1_values, x2_values = aimd(1.2, 6.3, threshold=C, iterations=ITERATESMAX)
plot(x1_values, x2_values, C, chart_title="AIMD")
```

```python
x1_values_1, x2_values_1 = aimd(1.2, 6.3, alpha=1, beta=0.5, threshold=C, iterations=ITERATESMAX)
x1_values_2, x2_values_2 = aimd(1.2, 6.3, alpha=2, beta=0.5, threshold=C, iterations=ITERATESMAX)
x1_values_3, x2_values_3 = aimd(1.2, 6.3, alpha=0.1, beta=0.5, threshold=C, iterations=ITERATESMAX)

plt.figure(figsize=(6, 4))
plt.plot(x1_values_1, x2_values_1, label="alpha=1", linewidth=0.5)
plt.plot(x1_values_2, x2_values_2, label="alpha=2", linewidth=0.5)
plt.plot(x1_values_3, x2_values_3, label="alpha=0.1", linewidth=0.5)
plt.plot([0, C], [C, 0], label='Efficiency Line', linestyle='--', color='red')
plt.plot([0, C], [0, C], label='Fairness Line', linestyle='--', color='green')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title("Effect of alpha parameter")
upper_limit = max([x1_values_1.max(), x2_values_1.max(), x1_values_2.max(), x2_values_2.max(), C])
plt.xlim(0, int(upper_limit)+1)
plt.ylim(0, int(upper_limit)+1)
plt.legend()
plt.grid(True)
plt.show()
```

```python
x1_values_1, x2_values_1 = aimd(1.2, 6.3, alpha=1, beta=0.5, threshold=C, iterations=ITERATESMAX)
x1_values_2, x2_values_2 = aimd(1.2, 6.3, alpha=1, beta=0.9, threshold=C, iterations=ITERATESMAX)
x1_values_3, x2_values_3 = aimd(1.2, 6.3, alpha=1, beta=0.1, threshold=C, iterations=ITERATESMAX)

plt.figure(figsize=(6, 4))
plt.plot(x1_values_1, x2_values_1, label="beta=0.5", linewidth=0.5)
plt.plot(x1_values_2, x2_values_2, label="beta=0.9", linewidth=0.5)
plt.plot(x1_values_3, x2_values_3, label="beta=0.1", linewidth=0.5)
plt.plot([0, C], [C, 0], label='Efficiency Line', linestyle='--', color='red')
plt.plot([0, C], [0, C], label='Fairness Line', linestyle='--', color='green')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title("Effect of beta parameter")
upper_limit = max([x1_values_1.max(), x2_values_1.max(), x1_values_2.max(), x2_values_2.max(), C])
plt.xlim(0, int(upper_limit)+1)
plt.ylim(0, int(upper_limit)+1)
plt.legend()
plt.grid(True)
plt.show()
```

## DCTCP

```python
x1_values, x2_values = dctcp(1.2, 6.3,alpha=1, g=0.2,threshold=C, iterations=ITERATESMAX)
plot(x1_values, x2_values, C, chart_title="DCTCP")
```

## CUBIC

```python
x1_values, x2_values = cubic(1.2, 6.3,threshold=C, iterations=ITERATESMAX)
plot(x1_values, x2_values, C, chart_title="TCP CUBIC")
```
